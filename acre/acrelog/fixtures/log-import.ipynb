```python
# import python's module
import pandas as pd
import json

# load data
log_df = pd.read_csv("operation-log.csv")  
log_df
# _df stands for data frame. It is a common suffix to indicate variable type. 
```

```python
# create a list which we will add dictionaries later
user = [] # suffix s is common way to indicate a list (array)
# ! check this DK
for operator in log_df['operator'].unique(): # we iterate over the result from unique function
    user.append( # append new element to the list
        { # curly bracket indicate the starting of dictionary
            "operator": operator, # each key-value pair
        } # the end of dictionary
    )
operator_df = pd.DataFrame(user)
operator_df.to_dict()
operator_df.to_dict()['operator']
index_name = operator_df.to_dict()['operator']
name2pk = dict((v, k+1) for k, v in index_name.items())

log_df['operator'] = log_df['operator'].map(name2pk) 
```

```python
# create a list which we will add dictionaries later
field = [] # suffix s is common way to indicate a list (array)
# ! check this DK
for location in log_df['location'].unique(): # we iterate over the result from unique function
    field.append( # append new element to the list
        { # curly bracket indicate the starting of dictionary
            "location": location, # each key-value pair
        } # the end of dictionary
    )
location_df = pd.DataFrame(field)
location_df.to_dict()
location_df.to_dict()['location']
index_name = location_df.to_dict()['location']
name2pk = dict((v, k+1) for k, v in index_name.items())
print(name2pk)
log_df['location'] = log_df['location'].map(name2pk) 
```

```python
# create a list which we will add dictionaries later
event = [] # suffix s is common way to indicate a list (array)
# ! check this DK
for operation in log_df['operation'].unique(): # we iterate over the result from unique function
    event.append( # append new element to the list
        { # curly bracket indicate the starting of dictionary
            "operation": operation, # each key-value pair
        } # the end of dictionary
    )
event_df = pd.DataFrame(event)
event_df
event_df.to_dict()
event_df.to_dict()['operation']
index_name = event_df.to_dict()['operation']
name2pk = dict((v, k+1) for k, v in index_name.items())

log_df['operation'] = log_df['operation'].map(name2pk) 
pd.to_datetime(log_df['date'])
log_df['date'] = pd.to_datetime(log_df['date']).dt.strftime('%Y-%m-%d')
```

```python
log = []
for index, row in log_df.iterrows(): # we iterate over the dataframe row by row
    log.append( # append new element to the list
        { # curly bracket indicate the starting of dictionary
            "model": "acrelog.log",  # the class name
            "pk": index+1, # index starts at 0 while primary key starts at 1. 
            "fields": { # this is a dictionary of attributes
                "location": row['location'],
                "event": row['operation'],
                "operator": row['operator'],
                "date": row['date'],
                "note": row['note'],
            }
        } # the end of dictionary
    )

json.dumps(log)
# or write the JSON to the file directly 
with open('log-data.json', 'w') as fp:
    json.dump(log, fp=fp, indent=4)
```
